```python
# analyzing data before and after lockdown

before_lockdown = data2[(data2["Month"]>=1) & (data2["Month"]<=3)]
before_lockdown
```

```python
after_lockdown = data2[(data2["Month"]>=4) & (data2["Month"]<=10)]
after_lockdown
```

```python
unemp_before_lockdown = before_lockdown.groupby(before_lockdown['Region'])[[' Estimated Unemployment Rate (%)']].mean().reset_index()
unemp_before_lockdown.head(3)
```

```python
unemp_after_lockdown = after_lockdown.groupby(after_lockdown['Region'])[[' Estimated Unemployment Rate (%)']].mean().reset_index()
unemp_after_lockdown.head(3)
```

```python
unemp_before_lockdown["unemp_after_lockdown"] = unemp_after_lockdown[' Estimated Unemployment Rate (%)']
new_unemp = unemp_before_lockdown
```

```python
new_unemp.rename(columns={' Estimated Unemployment Rate (%)':'unemp_before_lockdown'}, inplace=True)
new_unemp.head()
```

```python
new_unemp['percent_change_in_unemp'] = round((new_unemp["unemp_after_lockdown"] - new_unemp["unemp_before_lockdown"])/new_unemp["unemp_before_lockdown"],2)
new_unemp.head()
```

```python
plot_per = new_unemp.sort_values('percent_change_in_unemp')
```

```python
fig = px.bar(plot_per, x='Region', y='percent_change_in_unemp', title='percentage change in Unemployment in each state after lockdown',template='ggplot2')
```

```python
fig.show()
```
